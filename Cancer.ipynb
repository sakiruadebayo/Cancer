```python
df_cancer.dropna(axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_cancer['diagnosis'] = le.fit_transform(df_cancer['diagnosis'])
df_cancer['diagnosis'] 
```

```python
X = df_cancer.drop(columns = ['diagnosis','id'], axis = 1)
y = df_cancer.diagnosis.values.reshape(-1,1)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=200, max_depth=3, random_state=0)
rf_clf.fit(X_train, y_train)
```

```python
rf_clf.predict(X_train)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(rf_clf, X_train, y_train, cv=5, scoring= 'accuracy')
acc = score.mean()
print(acc)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=4)
clf.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(15,10))
plot_tree(clf)
plt.show()
```

```python
clf.predict(X_train)
```

```python
clf_score = cross_val_score(clf, X_train, y_train, scoring = 'accuracy')
acc = clf_score.mean()
print(acc)
```

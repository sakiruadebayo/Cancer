## 3. Data encoding

```python
# Label encoding of the target which is a categorical variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_cancer['diagnosis'] = le.fit_transform(df_cancer['diagnosis'])
df_cancer['diagnosis'] 
```

## 4. Model building and Evaluation 

```python
# Independent variable
X = df_cancer.drop(columns = ['diagnosis','id'], axis = 1)    # All rows and colmns except id and Target features

# Dependent variable
y = df_cancer.diagnosis.values.reshape(-1,1)                  # Target feature only
```

```python
# Dimension of independent variable
X.shape
```

```python
# Dimension ofdependent variable
y.shape
```

```python
# Split data into training and testing sets of ration 80:20
# 20% of test size selected
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=0)
```

### 4a. Random Forest Classifier

```python
# Random forest classifier
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=200, max_depth=3, random_state=0)
rf_clf.fit(X_train, y_train)    # train the model
```

```python
# Predicting X test
y_pred = rf_clf.predict(X_test)
y_pred
```

```python
# check accuracy of prediction
rf_clf.score(X_test,y_test)
```

```python
# check cross validation score
from sklearn.model_selection import cross_val_score
score = cross_val_score(rf_clf, X_train, y_train, cv=5, scoring= 'accuracy')
acc = score.mean()
print(acc)
```

#### Classification report

```python
#Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix\n',cf_matrix)
```

False Positive 64

False Negative 3

True Negative 2

True Positice 45

#### Confusion Matrix Heatmap

```python
sns.heatmap(cf_matrix, annot= True, fmt='g', xticklabels= ['benign', 'malignant'],
           yticklabels= ['benign', 'malignant'])
plt.ylabel('Prediction'),
plt.xlabel('Actual'),
plt.title('Confusion Matrix')
plt.show()
```

## 4b. Decision Tree Classifier

```python
# Decsion tree classifier
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=4)
clf.fit(X_train, y_train)         # train the model
```

```python
# Decision tree plot
from sklearn.tree import plot_tree
plt.figure(figsize=(15,10))
plot_tree(clf)
plt.show()
```

```python
# Predicting X test
y_pred = clf.predict(X_test)
y_pred
```

```python
# checking accuracy of prediction
clf.score(X_test, y_test)
```

```python
# check cross validation score
clf_score = cross_val_score(clf, X_train, y_train, scoring = 'accuracy')
acc = clf_score.mean()
print(acc)
```
